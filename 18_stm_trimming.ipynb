```python
from langgraph.graph import StateGraph, START,END
from langchain_groq import ChatGroq
from dotenv import load_dotenv

from langgraph.checkpoint.memory import InMemorySaver
from langchain_core.messages.utils import trim_messages, count_tokens_approximately

from typing import TypedDict, Annotated
from langchain_core.messages import HumanMessage, BaseMessage
from langgraph.graph.message import add_messages
```

```python
load_dotenv()
```

```python
llm = ChatGroq(model='llama-3.3-70b-versatile')
```

```python
MAX_TOKENS = 150
```

```python
class ChatState(TypedDict):
    messages: Annotated[list[BaseMessage], add_messages]
```

```python
def call_mode(state: ChatState): # by default MessagesState
    messages = trim_messages(
        state['messages'],
        strategy='last',
        max_tokens=MAX_TOKENS,
        token_counter=count_tokens_approximately
                             )
    response = llm.invoke(messages)
    
    return {
        'messages': [response]
    }
```

```python
graph = StateGraph(ChatState)

graph.add_node('call_mode', call_mode)

graph.add_edge(START, 'call_mode')
```

```python
checkpointer = InMemorySaver()
graph = graph.compile(checkpointer=checkpointer)
```

```python
config = {'configurable':{'thread_id':'thread-1'}}
```

```python
result = graph.invoke({'messages':{'role':'user','content':'Hi my name is Wijad Ullah!'}}, config=config)
print(result['messages'][-1].content)
```
